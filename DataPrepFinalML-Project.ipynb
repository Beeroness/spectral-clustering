# Data Prep for Project TK

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
file_name = 'cereal.csv'
# [calories, protein, fat, sodium, fiber, carbo, sugars, potass, vitamins, rating]
feature_indices = [3, 4, 5, 6, 7, 8, 9, 10, 11, 15]

# load and select features
print("Step 1: Loading Data and Feature Selection ")

data_array = np.genfromtxt(
        file_name,
        delimiter=',',
        skip_header=1,
        usecols=feature_indices,
        filling_values=np.nan,
        dtype=np.float64,
        # converter to handle the specific -1 missing vlaue indicator
        converters={i: lambda s: np.nan if s.strip() == b'-1' else float(s) for i in range(len(feature_indices))}
    )
X_raw = data_array


nan_mask = np.isnan(X_raw)
for i in range(X_raw.shape[1]):
    col_data= X_raw[~nan_mask[:, i], i]
    median_val = np.median(col_data) if col_data.size > 0 else 0.0
    X_raw[nan_mask[:, i], i] = median_val
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_raw)

print(f"Data Ready - matrix X_scaled shape: {X_scaled.shape}")
              
         
```
